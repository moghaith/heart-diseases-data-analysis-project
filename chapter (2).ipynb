```python
import pandas as pd


# Create a new column to classify chest pain type 2 as a group and other types as another group
df['nonanginal_or_not'] = df['chest_pain'].apply(lambda x: 'Non-Anginal Chest Pain' if x == 2 else 'Other Chest Pain')

# Count the occurrences of presence and absence of heart disease for each chest pain group
result = df.groupby(['nonanginal_or_not', 'target']).size().unstack(fill_value=0)

# Print the result
print(result)

# Create count plot
sns.countplot(data=df, y='nonanginal_or_not', hue='target')

# Show plot
plt.title('Presence and Absence of Heart Disease by Chest Pain Type')
plt.ylabel(' ')
plt.xlabel('Count')
plt.legend(["Present", "Absent"], title='Presence of Heart Disease')

plt.box(False)
plt.grid(False)
ax = plt.gca()
plt.show()
```

```python
import numpy as np
from scipy.stats import ttest_ind

n_pres = len(df[df['target'] == 1])
n_abs = len(df[df['target'] == 0])

pres_age = df[df['target'] == 1]['age']
abs_age = df[df['target'] == 0]['age']

t_statistic, p_value = ttest_ind(abs_age, pres_age, alternative='greater')

print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("Null hypothesis rejected. Mean age of patients with heart disease is lower than the mean age of patients without heart disease.")
else:
    print("Null hypothesis cannot be rejected. Mean age of patients with heart disease is not signifcanly lower than the mean age of patients without heart disease.")


```

```python
import numpy as np
from scipy.stats import ttest_ind

n_pres = len(df[df['target'] == 1])
n_abs = len(df[df['target'] == 0])

pres_rest_bp = df[df['target'] == 1]['rest_bp']
abs_rest_bp = df[df['target'] == 0]['rest_bp']

t_statistic, p_value = ttest_ind(pres_rest_bp, abs_rest_bp, alternative= 'less')

print("T-statistic:", t_statistic)
print("P-value:", p_value)

if p_value < 0.05:
    print("Null hypothesis rejected. Mean blood pressure of patients with heart disease is lower than that of patients without heart disease.")
else:
    print("Null hypothesis cannot be rejected. Mean blood pressure  of patients with heart disease is not signifcanly lower than that of patients without heart disease.")
```
